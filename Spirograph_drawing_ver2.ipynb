## Spirograph with animation
Reference: <br>
http://www.mathematische-basteleien.de/spirographs.htm

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import clear_output
from IPython.display import Image

%matplotlib inline
```

```python
# variables
x = np.arange(0, 36000, 0.1)
theta = x * (np.pi / 180.0)
```

```python
# Parameters
R = 9.0
r = 5.0
a = 12.0
```

```python
# Draw one picture of "Hypocycloids"
plt.figure(figsize = (5, 5))
spg_xi = (R - r) * np.cos((r / R) * theta) + a * np.cos((1 - (r / R)) * theta)
spg_yi = (R - r) * np.sin((r / R) * theta) - a * np.sin((1 - (r / R)) * theta)

plt.scatter(spg_xi, spg_yi, marker = '.', s = 0.01, alpha = 0.5)
plt.axis('off')
plt.show()
```

```python
# Draw one picture of "Epicycloids"
plt.figure(figsize = (5, 5))
spg_xi = (R + r) * np.cos((r / R) * theta) - a * np.cos((1 + (r / R)) * theta)
spg_yi = (R + r) * np.sin((r / R) * theta) - a * np.sin((1 + (r / R)) * theta)

plt.scatter(spg_xi, spg_yi, marker = '.', s = 0.01, alpha = 0.5)
plt.axis('off')
plt.show()
```

```python
# Draw continuously changing parameters R, r, and a
for a in range(15, 20):
    for r in range(1, 20, 4):
        for R in range(5, 10):
            
                clear_output(True)
                plt.figure(figsize = (5, 5))

                spg_xi = (R - r) * np.cos((r / R) * theta) + a * np.cos((1 - (r / R)) * theta)
                spg_yi = (R - r) * np.sin((r / R) * theta) - a * np.sin((1 - (r / R)) * theta)

                plt.scatter(spg_xi, spg_yi, marker = '.', s = 0.01, alpha = 0.5)
                plt.axis('off')
                plt.show()
                print('          R =', R, ' r =', r, ' a =', a)
```

### Animation

```python
# Make GIF animation
ims = []
fig = plt.figure(figsize = (5, 5))

for a in range(20, 30, 1):
    for r in range(10, 15, 1):
        for R in range(20, 30, 1):
            
                plt.axis('off')
                spg_xi = (R - r) * np.cos((r / R) * theta) + a * np.cos((1 - (r / R)) * theta)
                spg_yi = (R - r) * np.sin((r / R) * theta) - a * np.sin((1 - (r / R)) * theta)
                
                im = plt.plot(spg_xi, spg_yi, linewidth = 0.5, alpha = 0.5)
                ims.append(im)
                
anim = animation.ArtistAnimation(fig, ims, interval = 100)
anim.save('spirograph_animation.gif', writer = 'pillow')
```
